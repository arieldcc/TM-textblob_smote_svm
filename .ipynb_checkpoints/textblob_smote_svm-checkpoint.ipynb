```python
def my_tokenizer(doc):
    doc = re.sub(r'@[A-Za-z0-9]+', '', doc)
    doc = re.sub(r'#[A-Za-z0-9]+', '', doc)
    doc = re.sub(r'RT[\s]', '', doc)
    doc = re.sub(r"http\S+", '', doc)
    doc = re.sub(r'[0-9]+', '', doc)
    doc = re.sub(r'(.)\1+',r'\1\1', doc)
    doc = re.sub(r'[\?\.\!]+(?=[\?.\!])', '',doc)
    doc = re.sub(r'[^a-zA-Z]',' ', doc)
    doc = re.sub(r'\b(\w+)( \1\b)+', r'\1', doc)
    doc = doc.replace('\n', ' ')
    doc = doc.translate(str.maketrans('', '', string.punctuation))
    doc = doc.strip(' ')   
    #Mengubah menjadi huruf kecil
    doc = doc.lower()    
    #Text Normalization
    doc = informal_to_formal_indo(doc)    
    #Punctuation Removal+Menghapus Angka
    doc = doc.translate(str.maketrans('', '', string.punctuation + string.digits))   
    #Whitespace Removal
    doc = doc.strip()
    #Tokenization
    doc = tokenizer.tokenize(doc)
    doc_token1 = [word for word in doc]
    #Stopwords Removal
    doc_token2 = [word for word in doc_token1 if word not in stopword.get_stopword()]   
    #Lemmatization
    doc_token3 = [lemmatizer.lemmatize(word) for word in doc_token2]    
    return doc_token3
```

# 3. Preprocessing
## Langkah 3.1: Pembersihan Teks (Cleansing)
Pada tahap ini, teks dibersihkan dari elemen-elemen yang tidak relevan, seperti mention (@), hashtag (#), URL, angka, dan tanda baca yang berlebihan.

```python
def cleansing(doc):
    # Menghapus mention (@username)
    doc = re.sub(r'@[A-Za-z0-9]+', '', doc)
    print("After removing mentions:", doc)
    
    # Menghapus hashtag (#hashtag)
    doc = re.sub(r'#[A-Za-z0-9]+', '', doc)
    print("After removing hashtags:", doc)
    
    # Menghapus RT (Retweet) dan karakter setelahnya
    doc = re.sub(r'RT[\s]', '', doc)
    print("After removing RT:", doc)
    
    # Menghapus URL (http://...)
    doc = re.sub(r"http\S+", '', doc)
    print("After removing URLs:", doc)
    
    # Menghapus angka
    doc = re.sub(r'[0-9]+', '', doc)
    print("After removing numbers:", doc)
    
    # Menghapus karakter yang berulang (misal: heeeellllloooo -> heloo)
    doc = re.sub(r'(.)\1+', r'\1\1', doc)
    print("After collapsing repeated characters:", doc)
    
    # Menghapus tanda tanya, titik, atau seru yang berulang
    doc = re.sub(r'[\?\.\!]+(?=[\?.\!])', '', doc)
    print("After removing multiple punctuation marks:", doc)
    
    # Menghapus karakter selain huruf
    doc = re.sub(r'[^a-zA-Z]', ' ', doc)
    print("After removing non-alphabet characters:", doc)
    
    # Menghapus kata yang muncul lebih dari sekali berturut-turut (misal: good good -> good)
    doc = re.sub(r'\b(\w+)( \1\b)+', r'\1', doc)
    print("After removing repeated words:", doc)
    
    return doc

data['data_cleansing'] = data['judul_berita'].apply(cleansing)
```

## Langkah 3.2: Case Folding
Pada tahap ini, seluruh teks diubah menjadi huruf kecil untuk keseragaman.

```python
def case_folding(doc):
    doc = doc.lower()  # Mengubah teks menjadi huruf kecil
    print("After case folding:", doc)
    return doc

data['data_case_folding'] = data['data_cleansing'].apply(case_folding)
```

## Langkah 3.3: Tokenisasi
Tokenisasi adalah proses memecah teks menjadi kata-kata (tokens).

```python
def tokenize(doc):
    doc = doc.split()
    print("After tokenization:", doc)
    return doc

data['tokens'] = data['data_case_folding'].apply(tokenize)
```

## Langkah 3.4: Penghapusan Stopwords
Pada tahap ini, kata-kata yang tidak memberikan banyak informasi, seperti "dan", "di", "ke", dll., akan dihapus.

```python
from nltk.corpus import stopwords

# Pastikan Anda sudah mengunduh daftar stopword untuk bahasa yang digunakan
nltk.download('stopwords')

def stopword_removal(tokens):
    stop_words = set(stopwords.words('indonesian'))  # Bisa sesuaikan dengan bahasa lain
    tokens_no_stopwords = [word for word in tokens if word not in stop_words]
    print("After stopwords removal:", tokens_no_stopwords)
    return tokens_no_stopwords

data['stopword_removal'] = data['tokens'].apply(stopword_removal)
```

## Langkah 3.5: Lematisasi
Lematisasi adalah proses mengubah kata ke bentuk dasar atau lemma-nya. Anda bisa menggunakan **WordNetLemmatizer** dari **NLTK**.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatization(tokens):
    doc_token3 = [lemmatizer.lemmatize(word) for word in tokens]  # Mengubah kata ke bentuk lemma
    print("After lemmatization:", doc_token3)
    return doc_token3

data['lemmatization'] = data['stopword_removal'].apply(lemmatization)
```

## Langkah 3.6: Text Normalization
Text normalization adalah mengubah kata-kata informal menjadi bentuk formal (misalnya, "gak" menjadi "tidak"). Biasanya, ini memerlukan sebuah kamus atau aturan yang didefinisikan. Anda bisa menggunakan fungsi informal_to_formal_indo yang Anda sebutkan sebelumnya, atau membuat kamus tersendiri.

```python
def informal_to_formal_indo(doc):
    informal_dict = {"gak": "tidak", "aja": "saja", "kamu": "anda"}
    # If doc is a list of tokens, join it into a string first
    if isinstance(doc, list):
        doc = " ".join(doc)
    doc = " ".join([informal_dict.get(word, word) for word in doc.split()])
    print("After text normalization:", doc)
    return doc

data['informal_to_formal_indo'] = data['lemmatization'].apply(informal_to_formal_indo)
```

```python
import pandas as pd  # For handling dataframes
from googletrans import Translator

# Initialize the translator
translator = Translator()

def translate_to_english(text):
    # Translate text to English
    try:
        translated = translator.translate(text, src='id', dest='en')  # 'id' is for Indonesian, 'en' for English
        return translated.text
    except Exception as e:
        print(f"Translation error: {e}")
        return text  # If there's an error, return the original text

# Assuming the 'informal_to_formal_indo' column contains the normalized text in Indonesian
data['informal_to_formal_indo_english'] = data['informal_to_formal_indo'].apply(translate_to_english)

# Display the updated dataframe using pandas' display method
import pandas as pd
from IPython.display import display

display(data)  # This will display the dataframe in Jupyter/Colab
```

# 5. Labeling data dengan textblob

```python
from textblob import TextBlob

# Sentiment analysis function
def get_sentiment(text):
    # Create a TextBlob object
    blob = TextBlob(text)
    # Get the polarity score
    polarity = blob.sentiment.polarity
    
    # Assign label based on polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Apply sentiment analysis to the 'informal_to_formal_indo_english' column
data['sentiment_label'] = data['informal_to_formal_indo_english'].apply(get_sentiment)

# Display the updated dataframe
display(data)  # This will display the dataframe in Jupyter/Colab
```
